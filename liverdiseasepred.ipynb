```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import f1_score, precision_score, recall_score, accuracy_score
import statsmodels.api as sm

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [3, 5, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

rf_model = RandomForestClassifier()

grid_search = GridSearchCV(estimator=rf_model, param_grid=param_grid, cv=3)
grid_search.fit(X_train, y_train)

best_params = grid_search.best_params_
best_model = grid_search.best_estimator_

y_pred = best_model.predict(X_test)

f1 = f1_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
accuracy = accuracy_score(y_test, y_pred)

X_train_sm = sm.add_constant(X_train)
logit_model = sm.Logit(y_train, X_train_sm)
result = logit_model.fit()

p_value = result.pvalues[1]

print("Best Parameters:", best_params)
print("F1 Score:", f1)
print("Precision:", precision)
print("Recall:", recall)
print("Accuracy:", accuracy)
print("P-value:", p_value)
```

```python
look_back = 15
```

```python
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator

train_generator = TimeseriesGenerator(Result_train, Result_train, length=look_back, batch_size=20)
test_generator = TimeseriesGenerator(Result_test, Result_test, length=look_back, batch_size=1)
```

```python
from keras.models import Sequential
from keras.layers import LSTM, Dense
```

```python
model = Sequential()
model.add( LSTM(10,
        activation='relu',
        input_shape=(look_back,1)))
```

```python
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
```

```python
num_epochs = 25
model.fit_generator(train_generator, epochs=num_epochs, verbose=1)
```

```python
prediction = model.predict_generator(test_generator)

Result_train = Result_train.reshape((-1))
Result_test = Result_test.reshape((-1))
prediction = prediction.reshape((-1))
```
